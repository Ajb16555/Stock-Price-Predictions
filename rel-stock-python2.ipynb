### Checking if the data is stationary

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    labels = ['ADF Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used']
    for value, label in zip(result, labels):
        print(f'{label}: {value}')
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary.")
    else:
        print("Weak evidence against the null hypothesis, fail to reject the null hypothesis. Data has a unit root and is non-stationary.")

```

```python
# Perform ADF test on the average price
print("ADF Test for Average Price:")
adf_test(data['Avg Price'].dropna())
```

### With KPSS test

```python
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
def kpss_test(series, **kw):
    statistic, p_value, n_lags, critical_values = kpss(series, **kw)
    print(f'KPSS Statistic: {statistic}')
    print(f'p-value: {p_value}')
    print(f'num lags: {n_lags}')
    print('Critical Values:')
    for key, value in critical_values.items():
        print(f'   {key} : {value}')
    if p_value < 0.05:
        print("Strong evidence against the null hypothesis, reject the null hypothesis. Data is not stationary.")
    else:
        print("Weak evidence against the null hypothesis, fail to reject the null hypothesis. Data is stationary.")

# Perform KPSS test on the average price
print("KPSS Test for Average Price:")
kpss_test(data['Avg Price'].dropna(), regression='c')
```

### ARIMA, SARIMAX and Winter-Holtz

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Define the stock symbol and the date range
symbol = 'RELIANCE.NS'
start_date = '2014-07-01'
end_date = '2024-06-30'

# Fetch the stock data
data = yf.download(symbol, start=start_date, end=end_date)

# Calculate the average price (Open, Low, High, Adj Close)
data['Avg Price'] = data[['Open', 'Low', 'High', 'Adj Close']].mean(axis=1)
```

```python
# Differencing to remove trend
data['Differenced'] = data['Avg Price'].diff().dropna()

# Drop NA values created by differencing for proper split
data = data.dropna(subset=['Differenced'])

# Split data into training and testing sets
train, test = train_test_split(data['Differenced'], test_size=0.2, shuffle=False)
```

```python
# Fit ARIMA model
arima_model = ARIMA(train, order=(1, 1, 1))  # Order (p, d, q) needs to be tuned
arima_fitted_model = arima_model.fit()

# Make predictions with ARIMA
arima_predictions = arima_fitted_model.forecast(steps=len(test))

# Forecast the next 30 days with ARIMA
arima_future_predictions = arima_fitted_model.get_forecast(steps=30)
arima_future_conf_int = arima_future_predictions.conf_int()
arima_future_dates = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')

# Fit SARIMAX model
sarimax_model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))  # Seasonal order (p, d, q, s) needs to be tuned
sarimax_fitted_model = sarimax_model.fit()

# Make predictions with SARIMAX
sarimax_predictions = sarimax_fitted_model.forecast(steps=len(test))

# Forecast the next 30 days with SARIMAX
sarimax_future_predictions = sarimax_fitted_model.get_forecast(steps=30)
sarimax_future_conf_int = sarimax_future_predictions.conf_int()
sarimax_future_dates = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')

# Fit Holt-Winters model
holt_winters_model = ExponentialSmoothing(train, trend='add', seasonal='add', seasonal_periods=12)  # Adjust parameters as needed
holt_winters_fitted_model = holt_winters_model.fit()

# Make predictions with Holt-Winters
holt_winters_predictions = holt_winters_fitted_model.forecast(steps=len(test))

# Forecast the next 30 days with Holt-Winters
holt_winters_future_predictions = holt_winters_fitted_model.forecast(steps=30)
holt_winters_future_dates = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')

```

```python
# Calculate evaluation metrics
arima_mae = mean_absolute_error(test, arima_predictions)
arima_mse = mean_squared_error(test, arima_predictions)
arima_rmse = np.sqrt(arima_mse)

sarimax_mae = mean_absolute_error(test, sarimax_predictions)
sarimax_mse = mean_squared_error(test, sarimax_predictions)
sarimax_rmse = np.sqrt(sarimax_mse)

holt_winters_mae = mean_absolute_error(test, holt_winters_predictions)
holt_winters_mse = mean_squared_error(test, holt_winters_predictions)
holt_winters_rmse = np.sqrt(holt_winters_mse)

print(f"ARIMA Model - MAE: {arima_mae}, MSE: {arima_mse}, RMSE: {arima_rmse}")
print(f"SARIMAX Model - MAE: {sarimax_mae}, MSE: {sarimax_mse}, RMSE: {sarimax_rmse}")
print(f"Holt-Winters Model - MAE: {holt_winters_mae}, MSE: {holt_winters_mse}, RMSE: {holt_winters_rmse}")
```

```python
# Plot the results
plt.figure(figsize=(14, 7))
plt.plot(data['Avg Price'], label='Original Data')
plt.plot(train.index, train, label='Train Data', color='blue')
plt.plot(test.index, test, label='Test Data', color='green')
plt.plot(test.index, arima_predictions, label='ARIMA Predictions', color='red')
plt.plot(test.index, sarimax_predictions, label='SARIMAX Predictions', color='orange')
plt.plot(test.index, holt_winters_predictions, label='Holt-Winters Predictions', color='purple')

```
